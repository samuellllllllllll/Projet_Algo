```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Définition des paramètres du problème
num_points = 10  # Nombre total de points
num_vehicles = 3  # Nombre total de camions
vehicle_capacity = 80  # Capacité maximale de chaque camion
num_generation = 1000 # Nombre d'itérations
population_size = 100 # Taille de la population
mutation_rate = 0.2 #Taux de mutation
crossover_rate = 0.8
capacities = [vehicle_capacity] * num_vehicles

# Générer des coordonnées aléatoires pour les points
points = np.random.rand(num_points, 2) * 100

# Définition des demandes pour chaque point de livraison
demands = 20

# Calculer la distance entre deux points
def distance(point1, point2):
    return np.linalg.norm(point1 - point2)

# Calculer la distance totale d'un trajet
def calculate_distance(route):
    total_distance = 0
    for i in range(len(route) - 1):
        total_distance += distance(points[route[i]], points[route[i+1]])
    return total_distance

# Calculer la demande totale d'un trajet
def calculate_demand(route):
    total_demand = 0
    for point in route:
        total_demand += demands[point]
    return total_demand

# Générer une population initiale aléatoire
def generate_initial_population():
    population = []
    for _ in range(population_size):
        route = list(range(1, num_points))
        random.shuffle(route)
        population.append(route)
    return population

# Sélectionner les meilleurs individus de la population
def selection(population, num_parents):
    distances = [calculate_distance(route) for route in population]
    indices = np.argsort(distances)
    return [population[i] for i in indices[:num_parents]]

# Croisement de deux parents pour produire des enfants
def crossover(parent1, parent2, crossover_rate):
    if random.random() <= crossover_rate:
        point = random.randint(1, len(parent1) - 1)
        child1 = parent1[:point] + [p for p in parent2 if p not in parent1[:point]]
        child2 = parent2[:point] + [p for p in parent1 if p not in parent2[:point]]
        return child1, child2
    else:
        return parent1, parent2

# Mutation d'un individu en échangeant deux points
def mutate(individual, mutation_rate):
    if random.random() <= mutation_rate:
        point1, point2 = random.sample(range(1, len(individual)), 2)
        individual[point1], individual[point2] = individual[point2], individual[point1]
    return individual


# Résoudre le CVRP avec un algorithme génétique
def solve_cvrp(crossover_rate, mutation_rate):
    population = generate_initial_population()

    for generation in range(num_generation):
        parents = selection(population, 10)
        offspring = []

        while len(offspring) < len(population) - len(parents):
            parent1, parent2 = random.choices(parents, k=2)
            child1, child2 = crossover(parent1, parent2, crossover_rate)
            offspring.append(mutate(child1, mutation_rate))
            offspring.append(mutate(child2, mutation_rate))

        population = parents + offspring

    best_route = min(population, key=calculate_distance)
    return best_route

# Obtenir les trajets pour chaque camion à partir de la meilleure solution
def get_trips(route):
    trips = [[] for _ in range(num_vehicles)]
    capacity_left = [vehicle_capacity] * num_vehicles

    for point in route:
        assigned = False
        for i in range(num_vehicles):
            if demands <= capacity_left[i]:
                trips[i].append(point)
                capacity_left[i] -= demands
                assigned = True
                break

        if not assigned:
            print(f"Le point de livraison {point} ne peut pas être affecté à un camion.")
            
        # Afficher la capacité restante du camion
    plt.text(points[point][0], points[point][1], f"Capacité restante : {capacities[i]}", fontsize=7, ha='center', va='center')
    return trips

# Résoudre le problème et afficher les résultats
best_route = solve_cvrp(crossover_rate, mutation_rate)
trips = get_trips(best_route)

colors = ['b', 'g', 'r', 'c', 'm', 'y']
start_point = 0  # Point de départ commun pour tous les camions

distances_camions = []

for i, trip in enumerate(trips):
    trip.insert(0, start_point)
    trip.append(start_point)
    x = [points[point][0] for point in trip]
    y = [points[point][1] for point in trip]
    plt.scatter(x, y, color=colors[i % len(colors)])
    plt.plot(x, y, color=colors[i % len(colors)])
    plt.text(points[start_point][0], points[start_point][1], f"Camion {i+1}", fontsize=9, ha='center', va='center')

    # Afficher la distance entre chaque point
    for j in range(len(trip) - 1):
        p1 = points[trip[j]]
        p2 = points[trip[j+1]]
        distance_str = f"{calculate_distance([trip[j], trip[j+1]]):.2f}"
        plt.text((p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2, distance_str, fontsize=7, ha='center', va='center')

    # Calculer et stocker la distance totale parcourue par le camion
    distance_camion = calculate_distance(trip)
    distances_camions.append(distance_camion)

# Afficher les distances totales parcourues par chaque camion
for i, distance in enumerate(distances_camions):
    print(f"Distance totale parcourue par le camion {i+1}: {distance:.2f}")

plt.scatter(points[:, 0], points[:, 1], color='k', marker='o')

# Légende
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=colors[i % len(colors)],
                              markersize=8, label=f'Camion {i+1}') for i in range(len(trips))]
plt.legend(handles=legend_elements)

plt.show()
```
