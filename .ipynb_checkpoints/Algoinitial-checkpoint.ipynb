```python
import random
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def generate_random_points(n):
    points = []
    for _ in range(n):
        x = random.randint(0, 100)  # Ajustez la plage au besoin
        y = random.randint(0, 100)  # Ajustez la plage au besoin
        point = (x, y)
        points.append(point)
    return points

def create_adjacency_matrix(points):
    n = len(points)
    adjacency_matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(i+1, n):
            # Calculer la distance euclidienne entre les points i et j
            distance = calculate_distance(points[i], points[j])
            # Arrondir la distance à deux décimales
            distance = round(distance, 2)
            # Définir la valeur de l'arête entre i et j dans la matrice d'adjacence
            adjacency_matrix[i][j] = distance
            adjacency_matrix[j][i] = distance
    return adjacency_matrix

def calculate_distance(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    distance = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
    return distance

# Générer 5 points aléatoires
random_points = generate_random_points(20)

# Créer la matrice d'adjacence
adjacency_matrix = create_adjacency_matrix(random_points)

# Créer le graphe à partir de la matrice d'adjacence
G = nx.from_numpy_array(adjacency_matrix)

# Tracer le graphe
pos = nx.spring_layout(G)  # Positionnement des noeuds
nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightblue', font_weight='bold', font_color='black', edge_color='none')

# Ajouter les pondérations des arêtes
#edge_labels = nx.get_edge_attributes(G, "weight")
#nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Choisir un sommet aléatoire comme point de départ
start_vertex = random.choice(range(len(random_points)))

# Mettre en évidence le sommet de départ
nx.draw_networkx_nodes(G, pos, nodelist=[start_vertex], node_color='red', node_size=500)

# Choisir des points de livraison aléatoires
delivery_vertices = random.sample([v for v in range(len(random_points)) if v != start_vertex], 8)  # Modifier le nombre de points de livraison (ici, 8)

# Ajouter un colis à chaque point de livraison
delivery_packages = {}
for vertex in delivery_vertices:
    capacity = random.randint(1, 10)  # Ajustez la plage de capacité des colis au besoin
    delivery_packages[vertex] = {"capacity": capacity}
    
# Mettre en évidence les points de livraison et afficher les informations sur les colis
for vertex, package in delivery_packages.items():
    nx.draw_networkx_nodes(G, pos, nodelist=[vertex], node_color='green', node_size=500)
    nx.draw_networkx_labels(G, pos, labels={vertex: f"\nCapacité : {package['capacity']}"}, font_color='black')

    
# Créer 3 camions avec une capacité de 30
num_trucks = 2
truck_capacity = 50
trucks = [{"capacity": truck_capacity, "current_capacity": 0, "current_location": start_vertex, "route": [start_vertex]} for _ in range(num_trucks)]


# Diviser les sommets à visiter entre les camions
remaining_vertices = [v for v in range(len(random_points)) if v != start_vertex]
vertices_per_truck = len(remaining_vertices) // num_trucks

# Répartir les sommets à visiter entre les camions
for i in range(num_trucks):
    truck = trucks[i]
    if i == num_trucks - 1:
        # Le dernier camion obtient tous les sommets restants
        truck["vertices_to_visit"] = remaining_vertices
    else:
        # Répartir équitablement les sommets entre les camions
        truck["vertices_to_visit"] = remaining_vertices[:vertices_per_truck]
        remaining_vertices = remaining_vertices[vertices_per_truck:]

    
# Générer une solution existante aléatoire pour chaque camion
for truck in trucks:
    remaining_vertices = truck["vertices_to_visit"]
    current_location = truck["current_location"]

    while remaining_vertices:
        next_location = random.choice(remaining_vertices)
        edge_weight = G[current_location][next_location]["weight"]

        # Vérifier si le prochain sommet a un colis et s'il y a suffisamment de capacité dans le camion
        if next_location in delivery_packages and truck["current_capacity"] + delivery_packages[next_location]["capacity"] <= truck["capacity"]:
            # Récupérer le colis
            truck["current_capacity"] += delivery_packages[next_location]["capacity"]

        truck["route"].append(next_location)
        truck["current_location"] = next_location
        remaining_vertices.remove(next_location)

        current_location = next_location

    truck["route"].append(truck["route"][0])
    
    
# Afficher les chemins des camions sur le graphe
colors = ['blue', 'purple', 'green', 'orange', 'brown']  # Couleurs des camions
for i, truck in enumerate(trucks):
    route = truck["route"]

    # Afficher le chemin du camion en couleur
    edge_colors = [colors[i]] * (len(route) - 1)
    edges_to_draw = [edge for edge, color in zip(zip(route, route[1:]), edge_colors) if color != 'none']
    nx.draw_networkx_edges(G, pos, edgelist=edges_to_draw, edge_color=colors[i], width=2.0)

    # Afficher le point de départ du camion en rouge
    start_location = route[0]
    nx.draw_networkx_nodes(G, pos, nodelist=[start_location], node_color='red', node_size=500)


# Afficher le graphe
plt.show()

# Afficher les routes des camions
for i, truck in enumerate(trucks):
    print(f"Camion {i+1} : {truck['route']}")
    
# Afficher la capacité restante dans chaque camion
for i, truck in enumerate(trucks):
    remaining_capacity = truck["capacity"] - truck["current_capacity"]
    print(f"Camion {i+1} - Capacité restante : {remaining_capacity}")
```

```python
# Calculer la fitness globale du problème (somme des valeurs des arêtes empruntées par tous les camions)
def calculate_global_fitness(G, trucks):
    global_fitness = 0
    for truck in trucks:
        route = truck["route"]
        for i in range(len(route) - 1):
            source = route[i]
            target = route[i + 1]
            edge_weight = G[source][target]["weight"]
            global_fitness += edge_weight
    return global_fitness

# Exemple d'utilisation
global_fitness = calculate_global_fitness(G, trucks)
print("Fitness globale du problème :", global_fitness)

```
